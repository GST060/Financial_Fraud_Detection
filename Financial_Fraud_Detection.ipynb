```python
# Convert Camel case to snake_case
df.columns = (
    df.columns
    .str.replace('([a-z0-9])([A-Z])', r'\1_\2', regex=True)
    .str.lower()
)
```

```python
df.columns
```

```python
df = df.rename(columns = {'ip address':'ip_address'})
```

```python
# Convert TransactionDate column beacuse above TransactionDate column is string type
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors = 'coerce')
```

# Make new metrics/ columns 

```python
# For Detect suspicious transaction timing 
df['transaction_hour'] = df['transaction_date'].dt.hour
```

```python
# For Detect abnormally high activity
df['transactions_per_account'] = df.groupby('account_id')['transaction_id'].transform('count')
```

```python
# For Detect Multiple devices 
df['device_per_account'] = df.groupby('account_id')['device_id'].transform('nunique')
```

```python
# For detect many location for one account
df['locations_per_account'] = df.groupby('account_id')['location'].transform('nunique')
```

```python
# For Detect Normal spending pattern
df['avg_transaction_amount'] = df.groupby('account_id')['transaction_amount'].transform('mean')
```

```python
# For detect time difference between tranction for detect rapid consecutive transactions.
df = df.sort_values(['account_id', 'transaction_date'])
df['time_diff'] = df.groupby('account_id')['transaction_date'].diff()
```
